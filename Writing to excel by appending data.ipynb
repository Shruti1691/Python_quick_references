### Code to populate data from Investing.com

```python
# pip install investpy
import investpy

#Date format is dd/mm/yyyy

StD = '03/12/2019'
EnD = '21/12/2021'

df = investpy.indices.get_index_historical_data('S&P 500 Real Estate', 'United States', StD,EnD ,
as_json=False, order='ascending', interval='Daily')

df
```

```python
# Dropping all the columns from dataframe except Closing price

newdf = df.drop(columns=['Open','High','Low','Volume','Currency'])
newdf
```

```python
# Dataframe for Dates
df1 = newdf.drop(columns=['Close'])
df1
```

```python
# Code to determine the next blank row in excel

import xlwings as xw
import xlwt 
from xlwt import Workbook

wb = xw.Book('ETF.xlsx')
X = wb.sheets['Sheet1'].range('A' + str(wb.sheets['Sheet1'].cells.last_cell.row)).end('up').row + 1
app = xw.apps.active
app.quit()
X
```

```python
# Open excel and write the close price from dataframe to the respective columns

from openpyxl import *
import pandas as pd
from openpyxl import load_workbook

book = load_workbook('ETF.xlsx')
writer = pd.ExcelWriter('ETF.xlsx', engine='openpyxl') 
writer.book = book

## For the code to work effectively do not change the excel file name and keep it in the same path as the notebook file

writer.sheets = dict((ws.title, ws) for ws in book.worksheets)
newdf.to_excel(writer, sheet_name='Sheet1', startcol=0,startrow=X, header=False)
writer.save()
writ
```
